# 📊 Mini-Projeto: Visualizações com Python + SQL (SQLite)

Este notebook demonstra como usar **Matplotlib** e **Pandas** para criar gráficos a partir de dados armazenados em um banco SQLite.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect("empresa_visual.db")
cursor = conn.cursor()
```

## 1. Criar tabela e inserir dados

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS funcionarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    cargo TEXT NOT NULL,
    salario REAL,
    departamento TEXT
)
""")

dados = [
    ("Ana", "Analista de Dados", 4500, "TI"),
    ("Bruno", "Cientista de Dados", 7000, "TI"),
    ("Carla", "Engenheira de Software", 8000, "TI"),
    ("Daniel", "Contador", 5000, "Financeiro"),
    ("Eva", "Assistente de RH", 3000, "RH")
]

cursor.executemany("INSERT INTO funcionarios (nome, cargo, salario, departamento) VALUES (?, ?, ?, ?)", dados)
conn.commit()
```

## 2. Consultar dados e carregar em Pandas

```python
df = pd.read_sql_query("SELECT * FROM funcionarios", conn)
df
```

## 3. Criar gráfico de salários por funcionário

```python
plt.figure(figsize=(8,5))
plt.bar(df['nome'], df['salario'], color='skyblue')
plt.title('Salário por Funcionário')
plt.xlabel('Funcionário')
plt.ylabel('Salário (R$)')
plt.show()
```

## 4. Criar gráfico de salário médio por departamento

```python
df_group = df.groupby('departamento')['salario'].mean()

plt.figure(figsize=(6,4))
df_group.plot(kind='bar', color='orange')
plt.title('Salário Médio por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Salário Médio (R$)')
plt.show()
```

## 5. Fechar conexão

```python
conn.close()
```
