# 🚀 Mini-Projeto Avançado: Python + SQL (SQLite)

Este notebook demonstra consultas SQL mais avançadas usando Python + SQLite.
Inclui uso de **JOIN**, **GROUP BY** e funções de agregação.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("empresa_avancado.db")
cursor = conn.cursor()
```

## 1. Criar tabelas relacionadas

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS departamentos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS funcionarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    cargo TEXT NOT NULL,
    salario REAL,
    departamento_id INTEGER,
    FOREIGN KEY(departamento_id) REFERENCES departamentos(id)
)
""")
```

## 2. Inserir dados em departamentos e funcionários

```python
departamentos = [("TI",), ("Financeiro",), ("RH",)]
cursor.executemany("INSERT INTO departamentos (nome) VALUES (?)", departamentos)

funcionarios = [
    ("Ana", "Analista de Dados", 4500, 1),
    ("Bruno", "Cientista de Dados", 7000, 1),
    ("Carla", "Engenheira de Software", 8000, 1),
    ("Daniel", "Contador", 5000, 2),
    ("Eva", "Assistente de RH", 3000, 3)
]
cursor.executemany("INSERT INTO funcionarios (nome, cargo, salario, departamento_id) VALUES (?, ?, ?, ?)", funcionarios)
conn.commit()
```

## 3. Exemplo de JOIN

```python
query = """
SELECT f.nome, f.cargo, d.nome AS departamento
FROM funcionarios f
INNER JOIN departamentos d ON f.departamento_id = d.id
"""
df_join = pd.read_sql_query(query, conn)
df_join
```

## 4. Exemplo de GROUP BY e agregação

```python
query = """
SELECT d.nome AS departamento, AVG(f.salario) AS salario_medio
FROM funcionarios f
INNER JOIN departamentos d ON f.departamento_id = d.id
GROUP BY d.nome
"""
df_group = pd.read_sql_query(query, conn)
df_group
```
