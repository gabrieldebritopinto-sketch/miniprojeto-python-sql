# 🚀 Mini-Projeto: Python + SQL (SQLite)

Este notebook demonstra como usar **Python** para criar e manipular um banco de dados **SQLite**.
Inclui criação de tabela, inserção de dados e consultas integradas com **Pandas**.

```python
import sqlite3
import pandas as pd
```

## 1. Criar conexão com banco SQLite

```python
conn = sqlite3.connect("empresa.db")
cursor = conn.cursor()
```

## 2. Criar tabela de funcionários

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS funcionarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    cargo TEXT NOT NULL,
    salario REAL
)
""")
```

## 3. Inserir dados na tabela

```python
dados = [
    ("Ana", "Analista de Dados", 4500),
    ("Bruno", "Cientista de Dados", 7000),
    ("Carla", "Engenheira de Software", 8000)
]

cursor.executemany("INSERT INTO funcionarios (nome, cargo, salario) VALUES (?, ?, ?)", dados)
conn.commit()
```

## 4. Consultar dados com SQL

```python
cursor.execute("SELECT * FROM funcionarios WHERE salario > 5000")
resultados = cursor.fetchall()

print("Funcionários com salário acima de 5000:")
for linha in resultados:
    print(linha)
```

## 5. Usar Pandas para ler dados do banco

```python
df = pd.read_sql_query("SELECT * FROM funcionarios", conn)
df
```

## 6. Fechar conexão

```python
conn.close()
```
